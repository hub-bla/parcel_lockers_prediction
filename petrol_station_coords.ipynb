```python
BASE_DATA_DIR = './data'
petrol_df = pd.read_excel(f"{BASE_DATA_DIR}/petrol_stations_in_poland.xls")
petrol_df = pd.DataFrame(petrol_df.iloc[3:].to_numpy(), columns=petrol_df.iloc[2].to_numpy())
```

```python
poznan_petrol_stations = petrol_df[petrol_df['MIEJSCOWOŚĆ']=='POZNAŃ']
poznan_petrol_stations
```

```python
from geopy.geocoders import Nominatim

# calling the Nominatim tool and create Nominatim class
loc = Nominatim(user_agent="Geopy Library")
lats = []
longs = []
for i, row in poznan_petrol_stations.iterrows():
    
    # enter the location name
    getLoc = loc.geocode(f"{row['ADRES']}, POZNAŃ")
    if getLoc is None:
        lats.append(None)
        longs.append(None)
        continue

    # print address
    print(getLoc.address)
    
    lats.append(getLoc.latitude)
    longs.append(getLoc.longitude)
    
    # print latitude and longitude
    print("Latitude = ", getLoc.latitude, "\n")
    print("Longitude = ", getLoc.longitude)
```

```python
poznan_petrol_stations
```

```python
poznan_petrol_stations.loc[:, 'longitude'] = pd.Series(longs)
poznan_petrol_stations.loc[:, 'latitude'] = pd.Series(lats)
```

```python
poznan_petrol_stations[poznan_petrol_stations['longitude'].notna()][['longitude', 'latitude', 'ADRES']]\
                    .to_csv(f"{BASE_DATA_DIR}/poznan_petrol_stations.csv", index=False)
```
