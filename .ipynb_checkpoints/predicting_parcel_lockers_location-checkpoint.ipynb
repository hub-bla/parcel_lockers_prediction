```python
import pandas as pd
import geopandas
import numpy as np
import matplotlib.pyplot as plt
```

```python
import json

def open_json(file_name):
    f = open(file_name, encoding="utf8")

    return json.load(f)


pharmacy = pd.DataFrame(open_json('pharmacy.json')['elements'])
pharmacy = geopandas.GeoDataFrame(
    pharmacy, geometry=geopandas.points_from_xy(pharmacy['lon'], pharmacy['lat']), crs='EPSG:4326'
)

shops = pd.DataFrame(open_json('shops.json')['elements'])
shops = geopandas.GeoDataFrame(
    shops, geometry=geopandas.points_from_xy(shops['lon'], shops['lat']), crs='EPSG:4326'
)
```

```python
parcel_lockers = open_json('lockers_poznan.json')['elements']
parcel_lockers = [{'lat': locker['lat'], 'lon': locker['lon'], 'brand': locker['tags'].get('brand', None)} for locker in parcel_lockers]
parcel_lockers = pd.DataFrame(parcel_lockers)
parcel_lockers = geopandas.GeoDataFrame(
    parcel_lockers, geometry=geopandas.points_from_xy(parcel_lockers['lon'], parcel_lockers['lat']), crs='EPSG:4326'
)
parcel_lockers = parcel_lockers.loc[parcel_lockers['brand'].notna()]
```

```python
parcel_lockers[parcel_lockers['brand'].str.contains('Allegro')]
```

```python
people_19_24 = geopandas.read_file('people_age_19_24.geojson')
people_19_24 = people_19_24.to_crs("epsg:32633")
people_mid_age = geopandas.read_file('people_mid_age.geojson')
people_mid_age = people_mid_age.to_crs("epsg:32633")
```

```python
people_19_24.plot(column='LICZBA', legend=True, figsize=(16,8))
```

```python
people_mid_age.plot(column='LICZBA', legend=True, cmap='inferno', figsize=(16,8))
```

```python
petrol_stations = pd.read_csv('poznan_petrol_stations.csv')
petrol_stations = geopandas.GeoDataFrame(
    petrol_stations, geometry=geopandas.points_from_xy(petrol_stations['longitude'], petrol_stations['latitude']), crs='EPSG:4326'
)
```

```python
population_density_grid = geopandas.read_file('population_density_500m.geojson')

population_density_grid = population_density_grid.to_crs('EPSG:4326')

population_density_grid.plot(column='LICZBA_MIESZKANCOW', legend=True, cmap='inferno')
```

```python
grid = population_density_grid['geometry']
lockers_in_grid_cell = grid.apply(lambda cell: cell.contains(parcel_lockers['geometry']))
population_density_grid['num_lockers']  = np.array([parcel_lockers[lockers_in_grid_cell.iloc[i]].shape[0] for i in range(grid.shape[0])])
population_density_grid['num_lockers'][population_density_grid['num_lockers'] > 0].max()
```

```python
petrol_stations_in_grid_cell = grid.apply(lambda cell: cell.contains(petrol_stations['geometry']))
population_density_grid['num_petrol_stations'] = np.array([petrol_stations[petrol_stations_in_grid_cell.iloc[i]].shape[0] for i in range(grid.shape[0])])
population_density_grid['num_petrol_stations'][population_density_grid['num_petrol_stations'] > 0].max()
```

```python
def num_in_grid(df, name):
    in_grid_cell = grid.apply(lambda cell: cell.contains(df['geometry']))
    population_density_grid[f'num_{name}'] = np.array([df[in_grid_cell.iloc[i]].shape[0] for i in range(grid.shape[0])])
```

```python
num_in_grid(pharmacy, 'pharmacy')
num_in_grid(shops, 'shops')
```

```python
population_density_grid.plot(column='num_lockers', legend=True, cmap='inferno')
```

```python
buildings = geopandas.read_file('buildings.geojson')
buildings = buildings.to_crs("epsg:32633")
buildings.plot(column='WYSOKOSC_BLOKU', legend=True, cmap='inferno')
```

```python
population_density_grid = population_density_grid.to_crs("epsg:32633")
population_density_grid['buildings_area_rate'] = population_density_grid['geometry'].apply(lambda cell: (cell.intersection(buildings['geometry']).area/cell.area).sum())
```

```python
people_mid_age_grid_matrix = population_density_grid['geometry'].apply(lambda cell: (cell.intersection(people_mid_age['geometry']).area/cell.area))

num_people_mid_age = people_mid_age['LICZBA'].to_numpy()
num_people_mid_age_in_grid_cell = []

for i, row in people_mid_age_grid_matrix.iterrows():
    num_people_mid_age_in_grid_cell.append((people_mid_age_grid_matrix.iloc[i] * people_mid_age['LICZBA'].to_numpy()).sum())
```

```python
people_19_24_grid_matrix = population_density_grid['geometry'].apply(lambda cell: (cell.intersection(people_19_24['geometry']).area/cell.area))
num_people_19_24 = people_19_24['LICZBA'].to_numpy()
num_people_19_24_in_grid_cell = []

for i, row in people_mid_age_grid_matrix.iterrows():
    num_people_19_24_in_grid_cell.append((people_19_24_grid_matrix.iloc[i] * people_19_24['LICZBA'].to_numpy()).sum())
```

```python
population_density_grid['people_mid_age'] = num_people_mid_age_in_grid_cell
population_density_grid['people_19_24'] = num_people_19_24_in_grid_cell
```

```python
population_density_grid[population_density_grid['num_lockers'] > 0]
```

```python
allegro_lockers_in_grid_cell = grid.apply(lambda cell: cell.contains(parcel_lockers.loc[parcel_lockers['brand'].str.contains('Allegro'),'geometry']))
population_density_grid['allegro_lockers']  = np.array([parcel_lockers[(parcel_lockers['brand'].str.contains('Allegro'))&(allegro_lockers_in_grid_cell.iloc[i])].shape[0] for i in range(grid.shape[0])])
population_density_grid['allegro_lockers'][population_density_grid['allegro_lockers'] > 0].max()
```

```python
population_density_grid['competition_lockers'] = population_density_grid['num_lockers'] - population_density_grid['allegro_lockers']
```

```python
train_df = population_density_grid[['LICZBA_MIESZKANCOW', 'num_pharmacy', 'num_shops', 'num_petrol_stations', 'buildings_area_rate',  'people_mid_age','competition_lockers', 'allegro_lockers']]
X = train_df.drop('allegro_lockers', axis=1)
X['people_mid_age'] = np.log1p(X['people_mid_age'])
y = train_df['allegro_lockers']
```

```python
import seaborn as sns

corr_matrix = train_df.corr()
mask = np.triu(corr_matrix.to_numpy())
sns.heatmap(corr_matrix, annot=True, mask=mask)
```

```python
from sklearn.model_selection import StratifiedKFold, KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor

lr = make_pipeline(StandardScaler(), LinearRegression(positive=True))

skf = StratifiedKFold(n_splits=5, random_state=1, shuffle=True)

scores = []
oof = np.zeros_like(y, dtype=float)
print()
for train_ids, test_ids in skf.split(X, y):
    X_train, y_train = X.iloc[train_ids], y.iloc[train_ids]
    X_test, y_test = X.iloc[test_ids], y.iloc[test_ids]

    lr.fit(X_train, y_train)

    predictions = lr.predict(X_test)
    score = mean_squared_error(y_test, predictions)
    oof[test_ids] = predictions - y_test
    scores.append(score)
    print(f"RMSE: {score}, Coefs: {lr[1].coef_}")


print(f"FINAL: MEAN: {np.mean(scores)}, STD: {np.std(scores)}")
```

```python
mask = oof > 0.9
population_density_grid.loc[mask,'potential_lockers'] = oof[mask]
population_density_grid.loc[population_density_grid['potential_lockers'].isna(), 'potential_lockers'] = 0
```

## Out of folds residuals -> potential places for new parcel lockers

```python
population_density_grid.plot(column='potential_lockers', legend=True, cmap='inferno')
```

```python
import folium
CENTER = (52.4, 16.9)

map = folium.Map(location=CENTER,zoom_start=11)
population_density_grid = population_density_grid.to_crs('EPSG:4326')
show_lockers_locations = True

for i, r in population_density_grid[(population_density_grid['potential_lockers']==1)| (population_density_grid['num_lockers']>0)].iterrows():
    if show_lockers_locations:
        points = parcel_lockers.loc[lockers_in_grid_cell.iloc[i], 'geometry'].to_numpy()
        
        for point in points:
            folium.Marker(location=[point.y, point.x]).add_to(map)
    sim_geo = geopandas.GeoSeries(r["geometry"]).simplify(tolerance=0.00001)
    geo_j = sim_geo.to_json()
    # print(geo_j)
    if r['potential_lockers'] > 0:
        geo_j = folium.GeoJson(data=geo_j, style_function=lambda x: {"fillColor": "orange"})
        folium.Popup(f"{r['potential_lockers']}").add_to(geo_j)
    else:
        geo_j = folium.GeoJson(data=geo_j, style_function=lambda x: {"fillColor": "blue"})
        folium.Popup(f"{r['num_lockers']}").add_to(geo_j)
    geo_j.add_to(map)
```
