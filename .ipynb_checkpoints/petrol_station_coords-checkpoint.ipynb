```python
petrol_df = pd.read_excel('petrol_stations_in_poland.xls')
```

```python
petrol_df = pd.DataFrame(petrol_df.iloc[3:].to_numpy(), columns=petrol_df.iloc[2].to_numpy())
```

```python
petrol_df
```

```python
!pip install geopy
```

```python
poznan_petrol_stations = petrol_df[petrol_df['MIEJSCOWOŚĆ']=='POZNAŃ']
poznan_petrol_stations
```

```python
# importing geopy library and Nominatim class
from geopy.geocoders import Nominatim

# calling the Nominatim tool and create Nominatim class
loc = Nominatim(user_agent="Geopy Library")
lats = []
longs = []
for i, row in poznan_petrol_stations.iterrows():
    
    # entering the location name
    getLoc = loc.geocode(f"{row['ADRES']}, POZNAŃ")
    if getLoc is None:
        lats.append(None)
        longs.append(None)
        continue
    print(getLoc.address)
    
    lats.append(getLoc.latitude)
    longs.append(getLoc.longitude)
    # printing address
    
    # printing latitude and longitude
    print("Latitude = ", getLoc.latitude, "\n")
    print("Longitude = ", getLoc.longitude)
```

```python
poznan_petrol_stations['longitude'] = longs
poznan_petrol_stations['latitude'] = lats
```

```python
poznan_petrol_stations[poznan_petrol_stations['longitude'].notna()][['longitude', 'latitude', 'ADRES']].to_csv('poznan_petrol_stations.csv', index=False)
```
